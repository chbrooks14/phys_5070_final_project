```python
## Import cell
%matplotlib inline

import numpy as np
import matplotlib.pyplot as plt
```

The point of this notebook is to show what range of initial conditions lead to bound orbits for the Henon-Heiles Potential. The Henon-Heiles Potential is defined as:

$V(r,\theta) = \frac{1}{2} r^2 + \frac{1}{3} r^3 sin(3 \theta)$

which in Cartesian coordinates is given by:

$V(x,y) = \frac{1}{2}x^2 + \frac{1}{2}y^2 +x^2y - \frac{1}{3}y^3$

```python
def V_HH(x, y):
    return 1/2*x**2 + 1/2*y**2 + x**2*y - 1/3*y**3

x = np.linspace(-2,2,1000)
y = np.linspace(-2,2,1000)
X, Y = np.meshgrid(x, y) # np.meshgrid is the recommended way of obtaining contour plots
Z = V_HH(X, Y)
plt.figure(figsize=(10,8))
CS1 = plt.contour(X, Y, Z, [.01, .05, .1, .5, 1, 2, 3, 4, 5], colors='k')
CS2 = plt.contour(X, Y, Z, [1/6], colors='r')
plt.clabel(CS1, inline=1, fontsize=8)
plt.clabel(CS2, inline=1, fontsize=8)
plt.xlabel('x')
plt.ylabel('y')
plt.title('Contour Plot of Henon-Heiles Potential V(x,y)')
```

This contour plot demonstrates that x and y are _unbounded_ for energies greater than $V(x,y) = \frac{1}{6}$ (shown in red). The goal of this project is to study the dynamical and chaotic behavior of _bound_ orbits for this potential, so this illustrates that for my analysis of this system I need to restrict my initial conditions for $x_0, y_0, p_{0x}, p_{0y}$ to those where $E \leq \frac{1}{6}$.
